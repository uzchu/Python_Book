```python
import matplotlib.pyplot as plt
```

```python
plt.plot([10,20,30,40])
plt.show()
```

```python
plt.plot([1,2,3,4],[12,13,14,15])
plt.show()
```

```python
plt.title('plotting')
plt.plot([10,20,30,40])
plt.show()
```

```python
plt.title('legend')
plt.plot([1,2,3,4],label='asc')
plt.plot([4,3,2,1],label='desc')
plt.legend()
plt.show()
```

```python
plt.title('legend')
plt.plot([1,2,3,4],color='skyblue',label='asc')
plt.plot([4,3,2,1],'g',label='desc')
plt.legend()
plt.show()
```

```python
plt.title('legend')
plt.plot([1,2,3,4],color='skyblue',linestyle='--',label='asc')
plt.plot([4,3,2,1],'g',label='desc',ls=':')
plt.legend()
plt.show()
```

```python
plt.title('legend')
plt.plot([1,2,3,4],'r.--',label='asc') # ,,  
plt.plot([4,3,2,1],'g^',label='desc')
plt.legend()
plt.show()
```
