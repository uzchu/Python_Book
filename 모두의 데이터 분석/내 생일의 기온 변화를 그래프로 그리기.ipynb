```python
import csv
f = open('seoul.csv', encoding='utf-8')
data = csv.reader(f)
next(data)
result = []
for row in data:
    if ''.join(row[4:]) != '':
        result.append(float(''.join(row[4:])))
print(result)
```

row`[-1]`이 자꾸 list index out of range 에러가 발생함 (row[4]도)\
그래서 row`[4:]`(list형태)로 뽑고, ''.join(row[4:]) (string형태)로 변환

```python
print(len(result))
```

```python
import csv
import matplotlib.pyplot as plt
f = open('seoul.csv', encoding='utf-8')
data = csv.reader(f)
next(data)
result = []
for row in data:
    if ''.join(row[4:]) != '':
        result.append(float(''.join(row[4:])))
plt.figure(figsize=(20,2))
plt.plot(result,'r')
plt.show()
```
